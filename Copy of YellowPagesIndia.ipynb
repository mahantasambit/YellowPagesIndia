```python
origin_link = 'http://yellowpages.org.in/garments-suits-And-saree/16'
links = []
links.append(origin_link)
for p in range(2,16):
  next_link = origin_link + '/' + str(p)
  links.append(next_link)
links
```

```python
comp_name = []
location = []
contacts = []
href = []
i =1
for link in links:
  res = requests.get(link)
  if res.status_code != 200:
    break
  print(i)
  i +=1
  soup = BeautifulSoup(res.text)
  h3_s = soup.find_all("h3", {"class": "title"})
  for h3 in h3_s:
    comp_name.append(h3.text)
    a_s = h3.find_all('a')
    href.append(a_s[0]['href'])

  ul_s = soup.find_all("ul", {"class": "place-info"})
  for ul in ul_s:
    li_s = ul.find_all('li')
    location.append(li_s[0].text)
    contact = li_s[1].text
    if  'no' in contact:
      contacts.append('NA')
    else :
      contacts.append(contact)
```

```python
df = pd.DataFrame()
df['Company_Name'] = comp_name
df['Location'] = location
df['Phone'] = contacts
df['href'] = href
```
